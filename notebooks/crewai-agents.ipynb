## 1️⃣ Create Your First Agent

```python
from crewai import Agent, Task, Crew, Process

# Define a simple agent
writer = Agent(
    role="Content Writer",
    goal="Write clear, engaging technical content",
    backstory="You are an experienced tech writer who simplifies complex topics.",
    verbose=True,
)

# Define a task
writing_task = Task(
    description="Write a 200-word introduction to Retrieval-Augmented Generation (RAG).",
    expected_output="A clear, engaging introduction suitable for developers new to RAG.",
    agent=writer,
)

# Create and run the crew
crew = Crew(agents=[writer], tasks=[writing_task], verbose=True)
result = crew.kickoff()
print(result)
```

## 2️⃣ Multi-Agent Team

```python
# Research Agent
researcher = Agent(
    role="AI Research Analyst",
    goal="Find the most relevant and accurate information on a topic",
    backstory="You are a meticulous researcher who always cites sources and verifies facts.",
    verbose=True,
)

# Writer Agent  
writer = Agent(
    role="Technical Writer",
    goal="Transform research into polished, readable articles",
    backstory="You write technical content that developers love — clear, practical, with code examples.",
    verbose=True,
)

# Tasks (sequential — writer uses researcher's output)
research_task = Task(
    description="Research the current state of AI agents in 2025. Cover: major frameworks, trends, and predictions.",
    expected_output="A detailed research brief with key findings and sources.",
    agent=researcher,
)

writing_task = Task(
    description="Write a blog post based on the research findings. Include code examples where relevant.",
    expected_output="A 500-word blog post ready for publication.",
    agent=writer,
)

# Run the crew
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, writing_task],
    process=Process.sequential,
    verbose=True,
)

result = crew.kickoff()
print("\n" + "="*50)
print(result)
```

## 3️⃣ Agents with Tools

```python
from crewai.tools import tool

@tool("Word Counter")
def count_words(text: str) -> int:
    """Count the number of words in a given text."""
    return len(text.split())

@tool("Text Summarizer")
def summarize_key_points(text: str) -> str:
    """Extract key bullet points from text."""
    sentences = text.split('.')
    key_points = [s.strip() for s in sentences[:5] if len(s.strip()) > 20]
    return '\n'.join(f'• {p}' for p in key_points)

# Agent with tools
editor = Agent(
    role="Content Editor",
    goal="Review and improve content quality",
    backstory="You are a senior editor who ensures content meets quality standards.",
    tools=[count_words, summarize_key_points],
    verbose=True,
)

edit_task = Task(
    description="Review the following text: 'AI agents are autonomous systems that can perceive their environment, make decisions, and take actions. They use LLMs as their reasoning engine. Popular frameworks include CrewAI, AutoGen, and LangGraph. These frameworks enable developers to build complex multi-agent systems.' Count the words and extract key points.",
    expected_output="Word count and key bullet points from the text.",
    agent=editor,
)

crew = Crew(agents=[editor], tasks=[edit_task], verbose=True)
result = crew.kickoff()
print(result)
```

## 🎯 Key Takeaways

| Concept | Description |
|---------|-------------|
| **Agent** | Autonomous unit with role, goal, backstory, and tools |
| **Task** | Specific work item assigned to an agent |
| **Crew** | Team of agents working together |
| **Process** | How tasks are executed (sequential or hierarchical) |
| **Tools** | Custom functions agents can call |
