## 1️⃣ Quick Start — 5 Lines of RAG

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Document

# Create sample documents
documents = [
    Document(text="LlamaIndex is a data framework for LLM applications. It provides tools to ingest, structure, and query data for use with LLMs."),
    Document(text="RAG (Retrieval-Augmented Generation) combines retrieval of relevant information with LLM generation to produce accurate, grounded responses."),
    Document(text="Vector databases store embeddings and enable similarity search. Popular options include ChromaDB, Pinecone, Weaviate, and Qdrant."),
    Document(text="Chunking strategies include fixed-size, sentence-based, and semantic chunking. The choice affects retrieval quality significantly."),
    Document(text="Embedding models convert text into numerical vectors. OpenAI's text-embedding-3-small is a good default choice for most applications."),
]

# Build index and query — that's it!
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

response = query_engine.query("What is RAG?")
print(response)
```

## 2️⃣ Custom LLM and Embedding Model

```python
from llama_index.core import Settings
from llama_index.llms.openai import OpenAI
from llama_index.embeddings.openai import OpenAIEmbedding

# Configure globally
Settings.llm = OpenAI(model="gpt-4o-mini", temperature=0.2)
Settings.embed_model = OpenAIEmbedding(model="text-embedding-3-small")
Settings.chunk_size = 512
Settings.chunk_overlap = 50

# Rebuild index with new settings
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine(similarity_top_k=3)

response = query_engine.query("What are the best vector databases?")
print(response)
```

## 3️⃣ Chat Engine (Conversational RAG)

```python
# Convert to chat engine for multi-turn conversations
chat_engine = index.as_chat_engine(
    chat_mode="condense_plus_context",
    system_prompt="You are a helpful AI assistant that answers questions about RAG and LLM frameworks. Be concise.",
)

# First question
response = chat_engine.chat("What is LlamaIndex?")
print(f"Bot: {response}\n")

# Follow-up (uses conversation context)
response = chat_engine.chat("How does it compare to other frameworks?")
print(f"Bot: {response}")
```

## 4️⃣ View Retrieved Sources

```python
query_engine = index.as_query_engine(similarity_top_k=3)
response = query_engine.query("Tell me about embedding models and chunking")

print(f"Answer: {response}\n")
print("--- Sources ---")
for i, node in enumerate(response.source_nodes, 1):
    print(f"\n{i}. Score: {node.score:.4f}")
    print(f"   Text: {node.text[:100]}...")
```
