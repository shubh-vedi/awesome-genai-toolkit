## 1️⃣ Setup & First Prediction

```python
import dspy

# Configure the language model
lm = dspy.LM("openai/gpt-4o-mini")
dspy.configure(lm=lm)

# Simplest possible call — a "signature" defines input → output
qa = dspy.Predict("question -> answer")

result = qa(question="What is the capital of Japan?")
print(f"Answer: {result.answer}")
```

## 2️⃣ Signatures — The Core Concept

```python
# Different signatures for different tasks

# Summarization
summarizer = dspy.Predict("document -> summary")
result = summarizer(document="DSPy is a framework from Stanford NLP for programming with LLMs. Instead of writing prompts, you write signatures and modules, and DSPy compiles them into optimized prompts.")
print(f"Summary: {result.summary}\n")

# Classification
classifier = dspy.Predict("text -> label: str, confidence: float")
result = classifier(text="I absolutely loved this product! Best purchase ever!")
print(f"Label: {result.label}")
print(f"Confidence: {result.confidence}")
```

## 3️⃣ Chain of Thought — Automatic Reasoning

```python
# ChainOfThought adds reasoning steps automatically
cot = dspy.ChainOfThought("question -> answer")

result = cot(question="A bat and a ball cost $1.10 in total. The bat costs $1.00 more than the ball. How much does the ball cost?")

print(f"Reasoning: {result.reasoning}")
print(f"Answer: {result.answer}")
```

## 4️⃣ Building a Custom Module

```python
class ContentGenerator(dspy.Module):
    """Generate content with automatic quality check."""
    
    def __init__(self):
        super().__init__()
        self.generate = dspy.ChainOfThought("topic, audience -> content")
        self.evaluate = dspy.Predict("content, criteria -> score: float, feedback: str")
    
    def forward(self, topic, audience="developers"):
        # Generate content
        content = self.generate(topic=topic, audience=audience)
        
        # Self-evaluate
        evaluation = self.evaluate(
            content=content.content,
            criteria="clarity, accuracy, engagement, and relevance"
        )
        
        return dspy.Prediction(
            content=content.content,
            reasoning=content.reasoning,
            score=evaluation.score,
            feedback=evaluation.feedback
        )

# Use the module
generator = ContentGenerator()
result = generator(topic="Why RAG matters for production LLM apps")

print(f"Content: {result.content}\n")
print(f"Score: {result.score}")
print(f"Feedback: {result.feedback}")
```

## 🎯 DSPy vs Traditional Prompting

| Aspect | Traditional | DSPy |
|--------|------------|------|
| **Prompts** | Hand-written, fragile | Auto-compiled, optimized |
| **Iteration** | Manual trial and error | Metric-driven optimization |
| **Composability** | Copy-paste prompts | Modular, reusable programs |
| **Model switching** | Rewrite all prompts | Just change config |
